## Data Cleaning

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.head(20)
```

```python
countries = ['Kenya', 'Uganda', 'Tanzania', 'Nigeria', 'South Africa']
df = df[df['location'].isin(countries)]
```

```python
df.head(50)
```

```python
df.tail(20)
```

```python
# Drop rows with missing date or total_cases
df = df.dropna(subset=['date', 'total_cases'])
```

```python
# Convert date column
df['date'] = pd.to_datetime(df['date'])
```

```python
# Fill or interpolate missing numeric values
df.fillna(method='ffill', inplace=True)
```

```python
df.isnull().sum()
```

## Exploratory Data Analysis(EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot total cases over time
for country in countries:
    country_df = df[df['location'] == country]
    plt.plot(country_df['date'], country_df['total_cases'], label=country)

plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.show()
```

south africa had the highest no of covid cases followed by Kenya then nigeria, Uganda and finally Tanzania


```python
# Death rate
df['death_rate'] = df['total_deaths'] / df['total_cases']
```

```python
# Plot total vaccinations
for country in countries:
    country_df = df[df['location'] == country]
    plt.plot(country_df['date'], country_df['total_vaccinations'], label=country)

plt.title('Total Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Vaccinations')
plt.legend()
plt.show()
```

```python
import plotly.express as px

latest = df[df['date'] == df['date'].max()]
fig = px.choropleth(latest,
                    locations="iso_code",
                    color="total_cases",
                    hover_name="location",
                    color_continuous_scale="Viridis",
                    title="Total COVID-19 Cases by Country")
fig.show()
```

```python
# Select numeric columns only
numeric_df = df[['total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_vaccinations']]

# Optional: fill missing values (e.g., with 0 or forward fill)
numeric_df = numeric_df.fillna(0)
```

```python
# Create correlation matrix
corr_matrix = numeric_df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(10, 6))

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add title
plt.title('Correlation Heatmap of COVID-19 Metrics')
plt.show()
```
